```python
# Import required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style for plots
sns.set(style="whitegrid")

# Generate random data
np.random.seed(123)
data = np.random.normal(loc=70, scale=10, size=1000)

# Plot all visualizations
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Histogram
axes[0, 0].hist(data, bins=30, color='lightblue', edgecolor='black')
axes[0, 0].set_title("Histogram of Random Data")
axes[0, 0].set_xlabel("Values")
axes[0, 0].set_ylabel("Frequency")

# 2. Density Plot (Updated)
sns.kdeplot(data, color='blue', fill=True, ax=axes[0, 1])
axes[0, 1].set_title("Density Plot of Random Data")
axes[0, 1].set_xlabel("Values")
axes[0, 1].set_ylabel("Density")

# 3. Box Plot
sns.boxplot(data=data, color='lightgreen', ax=axes[1, 0])
axes[1, 0].set_title("Box Plot of Random Data")
axes[1, 0].set_ylabel("Values")

# 4. Violin Plot (Updated)
df = pd.DataFrame({
    'Category': np.random.choice(['A', 'B', 'C'], size=1000),
    'Values': np.random.normal(loc=70, scale=10, size=1000)
})
sns.violinplot(x='Category', y='Values', data=df, hue='Category', palette='pastel', ax=axes[1, 1], legend=False)
axes[1, 1].set_title("Violin Plot by Category")
axes[1, 1].set_xlabel("Category")
axes[1, 1].set_ylabel("Values")

# Show all plots
plt.tight_layout()
plt.show()
```
